# 時間物件

萬變不離其宗，所有商品的評價方法都可以以下列數學表達:
$$
    V = \sum^T_{t=1}\frac{CF_t}{(1+r)^{t}}
$$
其中時間 (t) 就扮演了一個重要的角色，因此我們有必要在設計評價套件時，統一設計出一套時間運算的規則。<br>
以下我們將會區分為五個項目，計算時間與其衍生的產物。

1. Date
2. Period
3. Daycount
4. Calendar
5. Schedule

## 1. Date

`Date` 作為最基礎的時間物件，其功能類似於 python `date` 模塊的 `date` 物件。<br>
然而 `date` 物件存在運算邏輯複雜的問題，因此為簡化邏輯，我們以之為基礎，設計出 `Date` 物件。<br>
安裝好 `valora` 套件後，我們可以使用以下代碼，建立(實例化) `Date` 物件。

```python
from valora.basic import Date

date1 = Date(2025, 1, 1)
```

我們可以使用簡單的加減法，對日期進行日期的調整。<br>
加法就是將日期向後移動若干日，減法則是將日期向前移動若干日，非常直覺!<br>

```python
date2 = date1 + 5
date3 = date1 - 5
print(f"{date1} 向後移動五日後為 {date2}")
print(f"{date1} 向前移動五日後為 {date3}")
```

此外，我們可以使用 `day_of_week`, `day_of_month`, `day_of_year` 這三種方法。<br>
分別找出該日期在當周、當月、當年的所在的日數。<br>

```python
print(f"{Date(2026, 4, 6)} 為星期 {Date(2026, 4, 6).day_of_week}")
print(f"{Date(2026, 4, 6)} 為4月的第 {Date(2026, 4, 6).day_of_month} 日")
print(f"{Date(2026, 4, 6)} 為2026年的第 {Date(2026, 4, 6).day_of_year} 日")
```

我們也可以使用 `begin_of_week`, `begin_of_month`, `begin_of_year` 找出該日期於當周、當月、當年的首日。<br>
想當然爾，也可用 `end_of_week`, `end_of_month`, `end_of_year` 找出該日期於當周、當月、當年的末日。<br>

```python
print(f"{Date(2026, 4, 6)} 於當周的首日為 {Date(2026, 4, 6).begin_of_week}")
print(f"{Date(2026, 4, 6)} 於當月的首日為 {Date(2026, 4, 6).begin_of_month}")
print(f"{Date(2026, 4, 6)} 於當年的首日為 {Date(2026, 4, 6).begin_of_year}")

print(f"{Date(2026, 4, 6)} 於當周的末日為 {Date(2026, 4, 6).end_of_week}")
print(f"{Date(2026, 4, 6)} 於當月的末日為 {Date(2026, 4, 6).end_of_month}")
print(f"{Date(2026, 4, 6)} 於當年的末日為 {Date(2026, 4, 6).end_of_year}")
```

使用 `is_leap` 可判斷當年度是否為閏年。

```python
print(f"2000 年是否為閏年: {Date.is_leap(2000)}")
print(f"2001 年是否為閏年: {Date.is_leap(2001)}")
```

使用 `date_range` 並給定起訖日，可提供起訖日內所有日期。

```python
beg_date = Date(2026, 7, 26)
end_date = Date(2026, 7, 29)
intervals = Date.date_range(beg_date, end_date)

print(f"{beg_date} 與 {end_date} 有以下日期: {intervals}")
```

提供以下 classmethod 方法，可快速自 `pd.Timestamp`, `datetime.datetime`, `str` 資料型態轉換為 `Date` 物件。<br>

```python
import pandas as pd
from datetime import date

t_stamp = pd.Timestamp(year=2026, month=3, day=14)
d_date = date(year=2026, month=3, day=14)
s_str = "20260314"

print(f"自 pd.Timestamp 轉換為 {Date.from_timestamp(t_stamp)}")
print(f"自 datetime.date 轉換為 {Date.from_datetime(t_stamp)}")
print(f"自 str 轉換為 {Date.from_string(s_str)}")
```

可透過比較運算子，進行日期間的比較。<br>

```python
date1 = Date(2026, 6, 26)
date2 = Date(2026, 7, 26)

print(f"{date1} 與 {date2} 相等: {date1 == date2}")
print(f"{date1} 與 {date2} 相異: {date1 != date2}")
print(f"{date1} 大於 {date2} : {date1 > date2}")
print(f"{date1} 大於等於 {date2} : {date1 >= date2}")
print(f"{date1} 小於 {date2} : {date1 < date2}")
print(f"{date1} 小於等於 {date2} : {date1 <= date2}")
```

## 2. Period

上述的 `Date` 物件已經足夠靈活，但多數商品條款，尤其是固定收益商品，常常會有固定期間給付利息的需求。<br>
因此，我們需要一個物件，再給定日期之下，去計算一段區間後的日期，這就是 `Period` 物件的用處。<br>

該物件需搭配 `PeriodType` 使用，共有以下 4 種區間，分別為:
1. DAILY
2. WEEKLY
3. MONTHLY
4. YEARLY

使用方法為
1.  `Date` + `Period` 為給定日期再加上特定期間。
2.  `Date` - `Period` 為給定日期再減去特定期間。

```python
from valora.basic import Period
from valora.basic import PeriodType

date1 = Date(2026, 12, 23)

print(f"{date1} 加上 3 周之日期為 {date1 + Period(PeriodType.WEEKLY, 3)}")
print(f"{date1} 加上 15 月之日期為 {date1 + Period(PeriodType.MONTHLY, 15)}")
print(f"{date1} 加上 100 年之日期為 {date1 + Period(PeriodType.YEARLY, 100)}")
```

## 3. DayCount

固定收益商品在計算每期給付利息時，多半使用以下數學表達:
$$
    P \times Rate \times \tau
$$
其中 $\tau$ 表示為計息區間的年化時間長度，考量到不同市場有不同的計息慣例，因此我們提供 `DayCount` 物件。<br>
該物件可透過 `year_fraction` 方法，再給定兩個日期之下，計算年化時間長度。<br>

目前提供的計息慣例有以下:
1. `ActActIsda`
2. `Act365Fixed`
3. `Act360`
4. `Thirty360US`
5. `ThirtyA360`
6. `ThirtyE360`

使用方法如下:

```python
from valora.basic import (
    ActActIsda,
    Act365Fixed,
    Act360,
    Thirty360US,
    ThirtyA360,
    ThirtyE360,
)

beg_date = Date(2005, 6, 11)
end_date = Date(2009, 6, 11)

print(f"當起訖日分別為 {beg_date} 與 {end_date}")
print(f"ActActIsda 標準下，年化時間差為 {ActActIsda.year_fraction(beg_date, end_date)}")
print(
    f"Act365Fixed 標準下，年化時間差為 {Act365Fixed.year_fraction(beg_date, end_date)}"
)
print(f"Act360 標準下，年化時間差為 {Act360.year_fraction(beg_date, end_date)}")
print(
    f"Thirty360US 標準下，年化時間差為 {Thirty360US.year_fraction(beg_date, end_date)}"
)
print(f"ThirtyA360 標準下，年化時間差為 {ThirtyA360.year_fraction(beg_date, end_date)}")
print(f"ThirtyE360 標準下，年化時間差為 {ThirtyE360.year_fraction(beg_date, end_date)}")
```

## 4. Calendar

營業日也是一個重要的日期衍生產物，所有交割、利率交付等現金流都僅發生在營業日，因此一個營業日日曆在評價中是必要的。<br>
提供營業日日曆有以下若干:

1. `FedWireCalendar`: 參考 Fed 營業日規則，以 rule-base 建立。
1. `Target2Calendar`: 參考 Target2 營業日規則，以 rule-base 建立。

```python
from valora.basic import FedWireCalendar, Target2Calendar

fed_calendar = FedWireCalendar()
tg2_calendar = Target2Calendar()
```

可透過 `is_business_day` 與 `is_holiday` 方法，再給定日期下，判斷該日是否為營業日。

```python
date1 = Date(2026, 1, 19)  # 馬丁路德紀念日

print(
    f"馬丁路德紀念日 {date1} 於 Fed 日曆中是否為假日: {fed_calendar.is_holiday(date1)}"
)
print(
    f"馬丁路德紀念日 {date1} 於 Target2 日曆中是否為營業日: {tg2_calendar.is_business_day(date1)}"
)
```

可透過 `add_holiday` 與 `remove_holiday` 針對日曆進行節假日的增減，以此達成客製化。

```python
date1 = Date(2026, 1, 19)  # 馬丁路德紀念日
date2 = Date(2026, 11, 27)  # 感恩節
date3 = Date(2026, 12, 25)  # 聖誕節
date4 = Date(2026, 12, 26)  # 聖誕節次日

added_tg2 = [date1, date2]
removed_tg2 = [date3, date4]
tg2_calendar.remove_holiday(removed_tg2)
tg2_calendar.add_holiday(added_tg2)

print(
    f"馬丁路德紀念日 {date1} 於 Target2 日曆中是否為節假日: {tg2_calendar.is_holiday(date1)}"
)
print(f"感恩節 {date2} 於 Target2 日曆中是否為節假日: {tg2_calendar.is_holiday(date2)}")
print(f"聖誕節 {date3} 於 Target2 日曆中是否為節假日: {tg2_calendar.is_holiday(date3)}")
print(
    f"聖誕節次日 {date4} 於 Target2 日曆中是否為節假日: {tg2_calendar.is_holiday(date4)}"
)
print("-" * 100)

added_fed = [date3, date4]
removed_fed = [date1, date2]
fed_calendar.remove_holiday(removed_fed)
fed_calendar.add_holiday(added_fed)

print(
    f"馬丁路德紀念日 {date1} 於 Fed 日曆中是否為節假日: {fed_calendar.is_holiday(date1)}"
)
print(f"感恩節 {date2} 於 Fed 日曆中是否為節假日: {fed_calendar.is_holiday(date2)}")
print(f"聖誕節 {date3} 於 Fed 日曆中是否為節假日: {fed_calendar.is_holiday(date3)}")
print(f"聖誕節次日 {date4} 於 Fed 日曆中是否為節假日: {fed_calendar.is_holiday(date4)}")
```

透過 `business_date_range` 可以再給定起訖日的條件下，衍生出當中所有營業日。

```python
beg_date = Date(2026, 1, 15)
end_date = Date(2026, 2, 2)
interval_bday = fed_calendar.business_date_range(
    beg_date,
    end_date,
    include_begin=True,  # 若起日為營業日，加入結果
    include_end=True,  # 若迄日為營業日，加入結果
)

print(f"起訖日為 {beg_date} 與 {end_date} 且依照 Fed 日曆規則(客製化後)")
print(f"營業日有: {interval_bday}")
```

固定收益商品於條款中會註明營業日調整慣例 (Business Day Convention)，以方便兩方確認若利息給付日遇到節假日時，<br>
該於何日進行交付，主要的慣例有 4 種:

1. FOLLOWING: 於下個營業日交付
2. PRECEDING: 於上個營業日交付
3. MODIFIED_FOLLOWING: 於下個營業日交付，但若下個營業日跨月時，改為上個營業日交付
4. MODIFIED_PRECEDING: 於上個營業日交付，但若上個營業日跨月時，改為下個營業日交付

我們提供 `BusinessDayConvention` 搭配 `adjust` 方法，可任意使用以上規則調整。

```python
from valora.basic.enum import BusinessDayConvention

date1 = Date(2026, 8, 1)

print(f"{date1} 為星期 {date1.day_of_week}")
print(
    f"FOLLOWING 調整後: {fed_calendar.adjust(date1, BusinessDayConvention.FOLLOWING)}"
)
print(
    f"PRECEDING 調整後: {fed_calendar.adjust(date1, BusinessDayConvention.PRECEDING)}"
)
print(
    f"MODIFIED_FOLLOWING 調整後: {fed_calendar.adjust(date1, BusinessDayConvention.MODIFIED_FOLLOWING)}"
)
print(
    f"MODIFIED_PRECEDING 調整後: {fed_calendar.adjust(date1, BusinessDayConvention.MODIFIED_PRECEDING)}"
)
```

透過 `begin_of_period` 方法，可以再指定日期與期間的條件下，找出該日期於該周、該月、該年的首個營業日。<br>
透過 `end_of_period` 方法，可以再指定日期與期間的條件下，找出該日期於該周、該月、該年的末個營業日。

```python
date1 = Date(2026, 8, 2)

print(f"{date1} 為星期 {date1.day_of_week}")
print(f"於當周的首個營業日: {fed_calendar.begin_of_period(date1, PeriodType.WEEKLY)}")
print(f"於當月的首個營業日: {fed_calendar.begin_of_period(date1, PeriodType.MONTHLY)}")
print(f"於當年的首個營業日: {fed_calendar.begin_of_period(date1, PeriodType.YEARLY)}")
print("-" * 100)
print(f"於當周的末個營業日: {fed_calendar.end_of_period(date1, PeriodType.WEEKLY)}")
print(f"於當月的末個營業日: {fed_calendar.end_of_period(date1, PeriodType.MONTHLY)}")
print(f"於當年的末個營業日: {fed_calendar.end_of_period(date1, PeriodType.YEARLY)}")
```

此外，部分固定收益商品可能存在 當月末日慣例 (end-of-month adjustment, eom)。<br>
該慣例為若商品生效日為當月最後一個交易日時，則後續各個計息區間起訖日都會貼其該月之末個營業日。<br>
因此我們可以使用 `is_eom` 再給定日期下，判斷該日是否為最後一個營業日。

```python
date1 = Date(2026, 10, 30)
date2 = Date(2026, 10, 31)

print(f"於 Fed 日曆下，{date1} 是否為當月最後的營業日: {fed_calendar.is_eom(date1)}")
print(f"於 Fed 日曆下，{date2} 是否為當月最後的營業日: {fed_calendar.is_eom(date2)}")
```

透過 `add_business_days` 我們可以再給定日期條件下，找出向前或向後 n 個交易日的日期。

```python
date1 = Date(2026, 10, 10)
n = 5

print(
    f"於 Fed 日曆下，{date1} 向前 {n} 營業日日期為 {fed_calendar.add_business_days(date1, -n)}"
)
print(
    f"於 Fed 日曆下，{date1} 向後 {n} 營業日日期為 {fed_calendar.add_business_days(date1, n)}"
)

# 2026/10/12 為哥倫布日
```

## 5. Schedule

有了 `Period` 與 `Calendar` 與 `DayCount` 後，我們可以算出債券的計息起訖日與付息日。<br>
透過 `Schedule` 我們可以快速統合前三者，讓我們更快生成日期的計劃表。<br>

```python
from valora.basic import Schedule

my_schedule = Schedule(
    beg_date=Date(2026, 12, 26),  # 起日
    end_date=Date(2027, 7, 13),  # 迄日
    tenor=Period(PeriodType.MONTHLY, 1),  # 生成間隔
    calendar=FedWireCalendar(),  # 營業日日曆
    convention=BusinessDayConvention.MODIFIED_FOLLOWING,  # 計息慣例
    forward_generate=False,  # 向前或向後生成
    eom_adjust=False,  # eom 調整
    include_begin=False,  # 是否包含起日
    include_end=True,  # 是否包含迄日
)

print(f"自 {Date(2026, 12, 26)} 到 {Date(2030, 12, 26)} 月頻率，")
print("且限制向前生成，使用 Fed 日曆跳過節假日情況下，")
print(f"生成日期為 {my_schedule.dates}")
```

透過 `append` 方法可以新增若干的日期於計畫表中。<br>
並且提供 `inplace` 功能，可以直接更新既有計畫表。<br>

```python
new_schedule = my_schedule.append(Date(2027, 5, 15))
print(f"新增 {Date(2027, 5, 15)} 於新計畫表中")
print(f"新計畫表日期: {new_schedule.dates}")

my_schedule.append(Date(2027, 1, 1), inplace=True)
print(f"新增 {Date(2027, 1, 1)} 於舊計畫表中")
print(f"舊計畫表日期: {my_schedule.dates}")
```

透過 `drop` 方法可以刪減若干日期於計畫表中，同樣具備 `inplace` 功能。

```python
my_schedule.drop(Date(2027, 1, 1), inplace=True)
print(f"刪減 {Date(2027, 1, 1)} 於舊計畫表中")
print(f"舊計畫表日期: {my_schedule.dates}")
```

透過 `shift` 方法可將所有日期向前或向後移動 n 個期間，請注意移動時不考慮營業日問題，同樣具備 `inplace` 功能。

```python
new_schedule = my_schedule.shift(period=Period(PeriodType.MONTHLY, -5), inplace=False)

print(f"原始日期: {my_schedule.dates}")
print(f"向前移動 5 月后日期: {new_schedule.dates}")
```

若 `shift` 後需要進行營業日調整，可透過 `adjust_to_business_day` 方法調整，同樣具備 `inplace` 功能。

```python
print(f"原始日期: {my_schedule.dates}")

new_schedule = my_schedule.shift(period=Period(PeriodType.MONTHLY, -5), inplace=False)

print(f"向前移動 5 月后日期: {new_schedule.dates}")

new_schedule = new_schedule.adjust_to_business_day(
    convention=BusinessDayConvention.MODIFIED_FOLLOWING,
    drop_duplicates=False,
    inplace=False,
)

print(f"向前移動 5 月且營業日調整後日期: {new_schedule.dates}")
```

透過 `clip` 可以指定上下界，並找出區間內的日期，且經由 `include_lower` 與 `include_upper`，<br>
可以決定是否要包含上下界，同樣具備 `inplace` 功能。

```python
new_schedule = my_schedule.clip(
    lower=Date(2027, 3, 1),
    upper=Date(2027, 5, 1),
    include_lower=True,
    inclide_upper=False,
)

print(f"原始日期: {my_schedule.dates}")
print(
    f"以 {Date(2027, 3, 1)} 到 {Date(2027, 5, 1)} 為區間擷取日期: {new_schedule.dates}"
)
```

當需要將每個日期都挪移到各週、各月、各年的最後一天時，可使用 `align_to_period_end` 方法。<br>
相對地，將每個日期都挪移到各週、各月、各年的第一天時，可使用 `align_to_period_begin` 方法。<br>
兩者皆提供 `inplace` 功能，移動時也皆不考慮營業日調整。<br>
並且考量可能有多個日期移動到同天的情況，可使用 `drop_duplicates` 決定是否保留。

```python
new_schedule = my_schedule.align_to_period_end(
    PeriodType.MONTHLY,
    drop_duplicates=False,
    inplace=False
)

print(f"原始日期: {my_schedule.dates}")
print(f"移動到月底並不刪除重複日期: {new_schedule.dates}")

new_schedule = my_schedule.align_to_period_begin(
    PeriodType.YEARLY,
    drop_duplicates=False,
    inplace=False
)

print(f"原始日期: {my_schedule.dates}")
print(f"移動到年初並不刪除重複日期: {new_schedule.dates}")
```

再計算債券時，常常需要計算各個計息區間的年化時間長度，可使用 `get_interval_year_fraction`。<br>
除此之外，我們也會計算生效日到各個付息日的年化時間長度，可使用 `get_cumulative_year_fraction`。

```python
from valora.basic import Act360
tau = my_schedule.get_interval_year_fraction(Act360)
c_tau = my_schedule.get_cumulative_year_fraction(Act360)

print(f"付息間隔日期為: {my_schedule.dates}")
print(f"在 Act360 ，年化時間長度為: {tau}")
print(f"在 Act360 ，累計年化時間長度為: {c_tau}")
```

以上含有 `inplace` 功能之方法，都可以使用鍊式規則完成多步操作。<br>
比如: 先在原日期計畫上，加入 2021-5-8，扣除 2027-3-15，整體向前移動 5 日，並且以 Following 調整營業日。

```python
new_schedule = (
    my_schedule
    .append(Date(2021,5,8))
    .drop(Date(2027,3,15))
    .shift(Period(PeriodType.DAILY, -5))
    .adjust_to_business_day(BusinessDayConvention.FOLLOWING)
)

print(f"原始日期: {my_schedule.dates}")
print(f"調整後日期: {new_schedule.dates}")
```
